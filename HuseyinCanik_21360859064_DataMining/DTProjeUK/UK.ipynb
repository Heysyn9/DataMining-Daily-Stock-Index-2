```python
import pandas as pd 
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.read_csv("London Stock Exchange Stock Price History.csv")

# Verileri hazrla
# 'Date' stununu datetime formatna evir
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# 'Price' stununu saysal formata evir (virglleri kaldrarak)
df['Price'] = pd.to_numeric(df['Price'].str.replace(',', ''), errors='coerce')

# Grafii iz
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Price'], color='blue', label='Fiyat')

# Tarih formatn ayarla
# X-ekseninde her yl gster
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Tarih etiketlerini dndr ve hizala
plt.gcf().autofmt_xdate()

# Grafii zelletir
plt.title('London Stock Exchanges Prices')
plt.xlabel('Tarih')
plt.ylabel('Fiyat')
plt.grid(True)
plt.legend()

# Grafii gster
plt.show()
```
