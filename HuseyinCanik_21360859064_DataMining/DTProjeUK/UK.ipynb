```python
import pandas as pd 
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.read_csv("London Stock Exchange Stock Price History.csv")

# Verileri hazırla
# 'Date' sütununu datetime formatına çevir
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# 'Price' sütununu sayısal formata çevir (virgülleri kaldırarak)
df['Price'] = pd.to_numeric(df['Price'].str.replace(',', ''), errors='coerce')

# Grafiği çiz
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Price'], color='blue', label='Fiyat')

# Tarih formatını ayarla
# X-ekseninde her yılı göster
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Tarih etiketlerini döndür ve hizala
plt.gcf().autofmt_xdate()

# Grafiği özelleştir
plt.title('London Stock Exchanges Prices')
plt.xlabel('Tarih')
plt.ylabel('Fiyat')
plt.grid(True)
plt.legend()

# Grafiği göster
plt.show()
```
